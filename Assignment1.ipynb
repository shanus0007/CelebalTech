```python
csv1=pd.read_csv("C:\\Users\\HP\\Downloads\\archive\\Combined_dataset.csv")
csv1
```

```python
# Remove all non-numeric characters
```

```python
csv1["final_price"] = (
    csv1["final_price"]
    .astype(str)
    .str.replace(r"[^0-9.]", "", regex=True)
)
```

```python
# Convert columns to numeric
```

```python
csv1["final_price"] = pd.to_numeric(
    csv1["final_price"],
    errors="coerce"
)
```

```python
csv1["ratings_count"] = pd.to_numeric(
    csv1["ratings_count"],
    errors="coerce"
)
```

```python
# Create derived column
```

```python
csv1["total_amount"] = (
    csv1["final_price"] * csv1["ratings_count"]
)
```

```python
print(csv1[["final_price", "ratings_count", "total_amount"]].head())
```
